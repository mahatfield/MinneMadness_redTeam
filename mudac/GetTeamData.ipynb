```python
import pandas as pd
```

```python
cbb = pd.read_html("https://www.sports-reference.com/cbb/seasons/2002-school-stats.html")[0]
```

```python
cbb = pd.DataFrame()
for i in range(2001,2020):
    link = "https://www.sports-reference.com/cbb/seasons/" + str(i) + "-school-stats.html"
    df = pd.read_html(link)[0]
    df.columns = df.columns.droplevel()
    df = df[~pd.isna(df['Rk']) & (df['Rk'] != "Rk")]
    df.columns.values[3] = 'W_overall'
    df.columns.values[4] = 'L_overall'
    df.drop(['Unnamed: 8_level_1','Unnamed: 8_level_1','Unnamed: 14_level_1','Unnamed: 17_level_1','Unnamed: 20_level_1','Unnamed: 11_level_1','W','L'], axis = 1,inplace = True)
    df = df[df['School'].str.contains('NCAA')]
    df['School'] = df['School'].str.replace('\xa0NCAA',"")
    df['Year'] = i
    cbb = pd.concat([cbb,df],axis = 0)
```

```python
cbb.drop(['Rk'], axis = 1,inplace = True)
cbb.to_csv("~/teams.csv",index = False)
```
